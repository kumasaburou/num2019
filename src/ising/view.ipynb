```python
init_spin = Ising.initialize([32,16], lowT=False)
spin = Ising(init_spin)
n = 1000
for i in range(n*10):
    if i % n == 0:
        sns.heatmap(spin.spin, vmin=-1, vmax=1)
        plt.show()
    Ising.metropolis(spin, 0.5)
```

```python
init = Ising.initialize([8,8])
spin = Ising(init)
mc = 100000
# spin.magnetization(1, mc, 'metropolis')
spin.equilibriate(mc, 3, 'metropolis')
```

```python
spin.magnetization(3, mc, 'metropolis')
```

```python
ts = np.linspace(1.5, 3.0)
mlist = []
for t in ts:
    spin.equilibriate(mc, t, 'metropolis')
    mlist.append(spin.magnetization(t, mc, 'metropolis'))
```
